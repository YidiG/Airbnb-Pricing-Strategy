```python
import pandas as pd
import numpy as np
df_1 = pd.read_csv("Raw_AirbnbData.csv")
```

```python
df_1.head(10)
```

```python
#dropping columns no used for data analysis 
df = df_1.drop(['id',"source","name", "bathrooms_text", "host_verifications","host_location","has_availability","host_neighbourhood", "listing_url","neighbourhood","calculated_host_listings_count","host_listings_count", "scrape_id",'last_scraped',"picture_url","host_url", "host_id",'description','neighborhood_overview','host_name', "host_about","host_thumbnail_url", "host_picture_url", "neighbourhood_group_cleansed","bathrooms","calculated_host_listings_count_entire_homes","calculated_host_listings_count_private_rooms","calculated_host_listings_count_shared_rooms",'first_review','last_review',"license","calendar_last_scraped","calendar_updated","minimum_minimum_nights","maximum_minimum_nights","minimum_maximum_nights","maximum_maximum_nights","minimum_nights_avg_ntm","maximum_nights_avg_ntm"], axis = 1) 
df.shape
```

```python
df.isnull().sum()
```

```python
#drop NA from the host since and review columns
df = df.dropna(subset=["host_since",'review_scores_rating','review_scores_accuracy','review_scores_cleanliness','review_scores_checkin','review_scores_communication','review_scores_location','review_scores_value'])
```

```python
#replace NA in host response time with no response
df['host_response_time'] = df['host_response_time'].replace(np.NaN,"no response")
df.host_response_time.unique()  
```

```python
df['host_response_time']
```

```python
#convert the percentage response rate into a float 
df['host_response_rate'] = df['host_response_rate'].str.rstrip("%").astype(float)/100 # Stripping % and converting to float
df['host_response_rate'] = df['host_response_rate'].replace(np.NaN,0) # Filling 0 for NaN
df.host_response_rate.unique()  
```

```python
#convert the percentage acceptance rate into a float 
df['host_acceptance_rate'] = df['host_acceptance_rate'].str.rstrip("%").astype(float)/100 # Stripping % and converting to float
df['host_acceptance_rate'] = df['host_acceptance_rate'].replace(np.NaN,0) # Filling 0 for NaN
df.host_acceptance_rate.unique()  
```

```python
#convert super host status to boolean
dictionary_replace = {'t': True, 'f': False}
boolean_host_is_superhost = df['host_is_superhost'].map(dictionary_replace)
df["host_is_superhost"] = boolean_host_is_superhost
df.host_is_superhost.unique()  
```

```python
#replace NA in host listing count to 1 - the minimum possible
df['host_total_listings_count'] = df['host_total_listings_count'].replace(np.nan,1.0)
df.host_total_listings_count.unique() 
```

```python
#convert host profile picture into a boolean
dictionary_replace = {'t': True, 'f': False}
boolean_host_has_profile_pic = df['host_has_profile_pic'].map(dictionary_replace)
df["host_has_profile_pic"] = boolean_host_has_profile_pic
df['host_has_profile_pic'] = df['host_has_profile_pic'].replace(np.nan,False)
df.host_has_profile_pic.unique() 
```

```python
#convert host identify verified to a boolean
dictionary_replace = {'t': True, 'f': False}
boolean_host_identity_verified = df['host_identity_verified'].map(dictionary_replace)
df["host_identity_verified"] = boolean_host_identity_verified
df['host_identity_verified'] = df['host_identity_verified'].replace(np.nan,"None")
df.host_identity_verified.unique() 
```

```python
#replace 0 accomidations to a minimum of 1 
df['accommodates'] = df['accommodates'].replace(0,1)
df.accommodates.unique()  
```

```python
df.isnull().sum()
```

```python
df.head(100)
```

```python
#convert price to type float and drop $ sign
df['price'] = df['price'].str.replace(',', '', regex=True)
df['price'] = df['price'].str.replace('$', '', regex=True)
df['price'].head(5)
```

```python
df["price"] = df['price'].apply(pd.to_numeric)
result = df.price.dtypes
print(result)
```

```python
df[(df.beds==0) & (df.bedrooms==0)]
df = df.drop(df[(df.beds==0) & (df.bedrooms==0)].index)
df.reset_index(inplace=True, drop=True)
df.shape
```

```python
#we will use price and accomodates to predict missing values for bedrooms
new_df =df[['price','accommodates',"bedrooms"]].copy()
new_df
```

```python
train_data = new_df.dropna()
train_data
```

```python
test_data = new_df[new_df["bedrooms"].isnull()]
test_data
```

```python
X_train = train_data.drop("bedrooms", axis = 1)
X_train
```

```python
Y_train = train_data["bedrooms"]
Y_train
```

```python
X_test = test_data.drop("bedrooms", axis = 1)
X_test
```

```python
Y_test = test_data["bedrooms"]
Y_test.unique()
```

```python
from sklearn import linear_model
LINREG = linear_model.LinearRegression()
LINREG.fit(X_train,Y_train)
```

```python
y_predict = LINREG.predict(X_test)
y_predictrounded = y_predict.round(0)
```

```python
X_test["predicted_bedrooms"] = y_predictrounded
X_test
```

```python
df["bedrooms"].fillna(X_test["predicted_bedrooms"], inplace=True)
```

```python
df.bedrooms.unique()
```
